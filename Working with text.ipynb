## The `.replace()` Method

```python
chi = pd.read_csv("chicago.csv").dropna(how = "all")
chi["Department"] = chi["Department"].astype("category")
chi.head(3)
```

```python
chi["Department"] = chi["Department"].str.replace("MGMNT","MANAGEMENT")
chi.head()
```

```python
chi["Employee Annual Salary"] = chi["Employee Annual Salary"].str.replace("$","").astype(float)
```

```python
chi.info()
```

```python
chi["Employee Annual Salary"].nlargest(5)
```

## Filtering with String Method

```python
chi = pd.read_csv("chicago.csv").dropna(how = "all")
chi["Department"] = chi["Department"].astype("category")
chi.head(3)
```

```python
mask = chi["Position Title"].str.lower().str.contains("water")
chi[mask]
```

```python
chi[chi["Position Title"].str.lower().str.startswith("water")]
```

```python
chi[chi["Position Title"].str.lower().str.endswith("ist")]
```

## More String Metods- `.strip()` ,`.rstrip()` and  `.lstrip()`

```python
chi = pd.read_csv("chicago.csv").dropna(how = "all")
chi["Department"] = chi["Department"].astype("category")
chi.head(3)
```

```python
chi["Name"] = chi["Name"].str.rstrip().str.rstrip()
```

```python
chi["Position Title"] = chi["Position Title"].str.strip()
```

## String Methods on Index and Columns

```python
chi = pd.read_csv("chicago.csv", index_col="Name").dropna(how = "all")
chi["Department"] = chi["Department"].astype("category")
chi.head(3)
```

```python
chi.index = chi.index.str.strip().str.title()
```

```python
chi.head()
```

```python
chi.columns = chi.columns.str.strip().str.upper()
```

```python
chi.head()
```

## Split String by Character with `.split()` Method

```python
chi = pd.read_csv("chicago.csv").dropna(how = "all")
chi["Department"] = chi["Department"].astype("category")
chi.head(3)
```

```python
chi["Name"].str.split(",").str.get(0).str.title().value_counts()
```

```python
chi["Position Title"].str.split(" ").str.get(0).value_counts()
```

## The expand and `n` Parameter of the `.split()` Method

```python
chi = pd.read_csv("chicago.csv").dropna(how = "all")
chi["Department"] = chi["Department"].astype("category")
chi.head(3)
```

```python
chi[["First Name","Last Name"]]=chi["Name"].str.split(",", expand=True)
chi.head()
```

```python
chi[["First Title W","Second title W"]] = chi["Position Title"].str.split(" ",expand=True, n = 1)
```
